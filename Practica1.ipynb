```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
a= np.array([3.1, 1, -0.5, -3.2, 6])
b= np.array([1, 3, 2.2, 5.1, 1])
```

Operaciones de multiplicación:

- multiplicación escalar se ejecuta como el producto punto por medio de la función np.dot() para dar como resultado un número, un escalar. 
    a.b= a1b1 + a2b2 + a3b3 + ...
- la función np.dot() sirve para realizar el producto punto entre vectores de igual tamaño, matriz.vector y matriz.matriz 

- El producto punto a punto coje cada elemento del vector y lo multiplica por el número que esté en la misma posición y el resultado es un vector del mismo tamaño.

```python
# Multiplicación escalar:
escalar = np.dot(a, b)
pap= a*b
print(f"""Resultado del producto escalar: {escalar:.2e} 
Resultado del producto punto a punto:  {pap}""")
```

Construcción de la matriz:

```python
A = np.array([[2, -1, -3], [4, 1.5, -2.5], [7.3, -0.9, 0.2]])
AT= np.transpose(A)
print(f"""Matriz: 
{A}

Matriz transpuesta: 
{AT}""")
```

## Comandos de la librería numpy:

- ### np.ones (shape, dtype=None, order='C', *, device=None, like=None)
devuelve un nuevo arreglo tipo numpy de solo unos según el tamaño indicado (https://numpy.org/doc/stable/reference/generated/numpy.ones.html)

- ### np.round (a, decimals=0, out=None)
Redondea uniformemente el número decimal a entero o a la cantidad de decimales específico. (https://numpy.org/doc/stable/reference/generated/numpy.round.html)

- ### np.ceil (x)
Es un redondeo forzado hacia arriba, al número entero inmediatamente mayor (https://numpy.org/doc/stable/reference/generated/numpy.ceil.html)

- ### np.floor (x)
Redondeo al número entero menor (https://numpy.org/doc/stable/reference/generated/numpy.floor.html)

```python
ones= np.ones((3,2))
round= np.round(5.123456789, decimals=2)
ceil1= np.ceil(5.1)
ceil2= np.ceil(-5.5)
floor1 = np.floor(5.1)
floor2 = np.floor(-5.5)

print(f"""Ejemplos:
numpy.ones:
{ones}

redondeo: {round}

ceil para núemro positivo (5.1): {ceil1}
ceil para número negativo (-5.5): {ceil2}

floor para número positivo (5.1): {floor1}
floor para número negativo (-5.5): {floor2}""")
```

```python
yn= np.sin(np.pi* 0.12* np.arange(101))  #Arange crea un arreglo con una secuencia de números igualmente espaciados.
y2n= np.cos(2*np.pi*0.03* np.arange(101))
sn= yn + y2n
tn= yn*y2n

# Señales originales
plt.figure(figsize=(15, 8))
plt.suptitle('Señales')
plt.subplot(2, 1, 1)
plt.plot(yn, label='y[n]', color='blue')
plt.plot(y2n, label='y2[n]', color='red')
plt.grid()
plt.title('Señales Y[n] y Y2[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.legend()  

plt.subplot(2, 1, 2)
plt.plot(sn, label='s[n]', color='green')
plt.plot(tn, label='t[n]', color='orange')
plt.grid()
plt.title('Señales S[n] y T[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.legend()  

plt.tight_layout()   # tight_layout arregla el espaciado entre gráficos para que no se solapen
plt.show()
```

## Repaso Pandas

```python
def estadisticas_notas(diccionario):
    # Convertir el diccionario en una Serie
    serie = pd.Series(diccionario)
    
    # Calcular estadísticas
    resultado = pd.Series({                     #convierte claves en índices y valores en datos.
        "mínima": serie.min(),                  # retorna el valor mínimo de la serie.
        "máxima": serie.max(),                  # retorna el valor máximo de la serie.
        "media": serie.mean(),                  # retorna el valor promedio de la serie
        "desviación típica": serie.std()        # retorna la desviación estándar de la serie.
    })
    
    return resultado

# Ejemplo:
notas = {
    "Mateo": 4.5,
    "Juan": 2,
    "Camila": 3.3,
    "Luis": 4,
    "Carolina": 1.8
}

print(estadisticas_notas(notas))
```

```python
# importar los datos

df= pd.read_csv("datos.csv", sep=";") #Se utiliza sep= ";" porque el separador de los datos es punto y coma.
df.head()  #Muestra las primeras filas del DataFrame datos para conocer la estructura de los datos.
```

```python
# Borrar la columna "Unnamed: 0" con la función drop
df = df.drop(columns=["Unnamed: 0"])
df.head()
```

```python
# cálculo del IMC

altura= df["Height"] / 100  # Convertir altura de cm a metros
peso= df["Weight"]

def IMC(peso, altura):
    df["BMI"]= peso/ (altura**2)
    #Agregar la columna de IMC al DataFrame
    return df

# Agregar la columna de IMC al DataFrame
df = IMC(peso, altura)
df.head()
```

```python
# Punto opcional: categorizar según el IMC en obesidad, sobrepeso, normal y bajo peso.
def clasificar(df):
    
    def clasificacion(valor):
        if valor < 18.5:
            return 'Bajo peso'
        elif valor < 25:
            return 'Normal'
        elif valor < 30:
            return 'Sobrepeso'
        else:
            return 'Obesidad'
    
    df['Categoria_IMC'] = df['BMI'].apply(clasificacion) #Selecciona la columna BMI y aplica la función clasificación
    return df

df = clasificar(df)
print(df[['Categoria_IMC','BMI','Weight','Height']].head())
```
