Quiz 1. teoría de modelos laboratorio
comparar el nivel de concentración de 2 voluntarios
- parámetro de entrada 1: Amplitud señal voluntario 1
- parámetro de entrada 2: Frecuencia señal voluntario 1
- parámetro de entrada 3: Amplitud señal voluntario 2
- parámetro de entrada 4: Frecuencia señal voluntario 2

Tiempo total del análisis: 25s 
Frecuencia de muestreo: 5 muestras/seg

```python
import numpy as np
import matplotlib.pyplot as plt

Periodo_muestreo=1/5 #segundos
TiempoTot=25 #segundos

#Creación del vector tiempo
t=np.arange(0,TiempoTot+Periodo_muestreo,Periodo_muestreo)
print(f"El vector tiempo tienen un total de {len(t)} datos")

def estudio(A1,F1,A2,F2):
    #Creación de las señales de los voluntarios
    #Señal sen= Asin(2pi*F*t)
    Voluntario1= A1*np.sin(2*np.pi*F1*t)
    Voluntario2= A2*np.sin(2*np.pi*F2*t)

    X116=[]
    X114=[]

    for i in range (len(Voluntario1)):
        if Voluntario1[i]>1.5:
            V1= X116.append(np.where(Voluntario1==Voluntario1[i])[0][0])  
        if Voluntario2[i]>1.5:
            V2= X114.append(np.where(Voluntario2==Voluntario2[i])[0][0])  
    
    
    M= np.column_stack((Voluntario1,Voluntario2))

    plt.figure()
    plt.plot(t,Voluntario1,label="Voluntario 1",color="red")
    plt.plot(t,Voluntario2,label="Voluntario 2",color="blue")
    plt.scatter(t[X116],Voluntario1[X116],color="black",label="Coincidencias V1")
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Nivel concentración")
    plt.bar_label 
    plt.show()
    return (np.shape(M))

      
estudio(4,9,2,4)
```
